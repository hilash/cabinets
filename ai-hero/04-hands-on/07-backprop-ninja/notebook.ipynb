```python
import torch
import torch.nn.functional as F

# Helper to compare manual vs. autograd
def cmp(name, dt, t):
    ex = torch.all(dt == t.grad).item()
    app = torch.allclose(dt, t.grad)
    maxdiff = (dt - t.grad).abs().max().item()
    print(f'{name:15s} | exact: {str(ex):5s} | approx: {str(app):5s} | maxdiff: {maxdiff:.4e}')
```
