```python
import torch
import torch.nn.functional as F
import random

words = open('names.txt', 'r').read().splitlines()
random.seed(42)
random.shuffle(words)

chars = sorted(list(set(''.join(words))))
stoi = {s: i + 1 for i, s in enumerate(chars)}
stoi['.'] = 0
itos = {i: s for s, i in stoi.items()}
V = len(stoi)
print('vocab size:', V)
```

## Build trigram (ctx1, ctx2) → target tensors

```python
def build(words):
    xs1, xs2, ys = [], [], []
    for w in words:
        chs = ['.', '.'] + list(w) + ['.']
        for a, b, c in zip(chs, chs[1:], chs[2:]):
            xs1.append(stoi[a])
            xs2.append(stoi[b])
            ys.append(stoi[c])
    return torch.tensor(xs1), torch.tensor(xs2), torch.tensor(ys)

n = len(words)
train_words = words[:int(0.8*n)]
val_words   = words[int(0.8*n):int(0.9*n)]
test_words  = words[int(0.9*n):]

Xtr1, Xtr2, Ytr = build(train_words)
Xva1, Xva2, Yva = build(val_words)
print('train bigrams:', Xtr1.shape[0])
```
